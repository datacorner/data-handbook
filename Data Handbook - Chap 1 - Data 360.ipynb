# Distributions

## Data distribution (bar chart)

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
data = [1, 2, 3, 2, 1, 4, 5, 4, 2, 1]
x = [1, 2, 3, 4, 5]
y = [data.count(val) for val in x]
plt.bar(x, y)
plt.title('Value Distribution', fontsize=10)
plt.show()
```

## Normal distribution

```python
from scipy.integrate import quad
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np

plt.figure(figsize=(20,10))

x_min, x_max = -4.0, 4.0
mean = 0.0
std = 1.0

x = np.linspace(x_min, x_max, 100)
y = scipy.stats.norm.pdf(x, mean, std)
plt.plot(x,y, color='red')

plt.xlim(x_min,x_max)
plt.ylim(0, 0.45)
plt.title('Normal distribution',fontsize=10)
plt.show()
```

**Normal distribution with standard deviation varying**

```python
from scipy.integrate import quad
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np

plt.figure(figsize=(20,10))

def TraceLoiNormale(std, mean, color):
  x_min, x_max = -4.0, 4.0
  mean = mean
  std = std

  x = np.linspace(x_min, x_max, 100)
  y = scipy.stats.norm.pdf(x, mean, std)
  plt.plot(x,y, color=color)

  plt.xlim(x_min,x_max)
  plt.ylim(0, 0.45)
  plt.title('Normal distribution ',fontsize=10)

TraceLoiNormale(1.0, 0.0, 'red')
TraceLoiNormale(2.0, 0.0, 'blue')
TraceLoiNormale(3.0, 0.0, 'green')

plt.show()
```
